```python
class ModeloIA:
    def __init__(self, nombre, precision_simulada=0.0): #Agregamos un nuevo parámetro opcional precision_simulada al constructor de la clase ModeloIA, que permite establecer una precisión simulada para el modelo, con un valor predeterminado de 0.0 si no se proporciona.
        self.nombre = nombre
        self.entrenado = False #Inicializamos un atributo entrenado en False para indicar que el modelo no ha sido entrenado aún.
        self.precision_simulada = precision_simulada #Asignamos el valor del parámetro precision_simulada al atributo de la instancia, lo que permitirá simular una precisión específica para el modelo durante las pruebas y evaluaciones.
    def entrenar(self, datos):
        print(f"[{self.nombre}] Iniciando entrenamiento.") #Agregamos un mensaje indicando que el entrenamiento del modelo ha comenzado, lo que proporciona retroalimentación visual sobre el proceso de entrenamiento.
        print(f"[{self.nombre}] Datos: {len(datos)} muestras") #Agregamos un mensaje que muestra la cantidad de muestras de datos utilizadas para entrenar el modelo, lo que ayuda a contextualizar el proceso de entrenamiento y su posible impacto en la precisión del modelo.
        self.entrenado = True #Marcamos el modelo como entrenado estableciendo el atributo entrenado en True, lo que permitirá que el modelo pueda realizar predicciones posteriormente.
        if self.precision_simulada == 0.0: #Si no se ha establecido una precisión simulada específica al crear la instancia del modelo, asignamos un valor predeterminado de 0.75 para simular una precisión razonable durante las pruebas y evaluaciones, lo que permite observar cómo el modelo se comporta con una precisión simulada sin necesidad de realizar un entrenamiento real.
            self.precision_simulada = 0.75 #Asignamos un valor predeterminado de 0.75 a precision_simulada si no se ha establecido previamente, lo que simula una precisión razonable para el modelo durante las pruebas y evaluaciones, permitiendo observar cómo el modelo se comporta con esta precisión simulada sin necesidad de realizar un entrenamiento real.
        print(f"[{self.nombre}] Precision: {self.precision_simulada:.2%}") #Agregamos un mensaje que muestra la precisión simulada del modelo después de entrenarlo, formateada como un porcentaje con dos decimales, lo que proporciona información sobre el rendimiento simulado del modelo basado en la precisión establecida.
    def predecir(self, entrada): #Agregamos un método predecir a la clase ModeloIA que toma una entrada y simula una predicción basada en la precisión simulada del modelo, lo que permite observar cómo el modelo genera predicciones de manera diferente según su precisión simulada establecida.
        if not self.entrenado:#Si el modelo no ha sido entrenado, lanzamos una excepción para indicar que no se pueden realizar predicciones, lo que refuerza la necesidad de entrenar el modelo antes de usarlo para generar predicciones.
            raise Exception("Modelo no entrenado") #Lanzamos una excepción si el modelo no ha sido entrenado para indicar que no se pueden realizar predicciones, lo que refuerza la necesidad de entrenar el modelo antes de usarlo para generar predicciones.
        print(f"[{self.nombre}] Prediccion base para: {entrada}") #Agregamos un mensaje que muestra la entrada para la cual se está generando una predicción, lo que proporciona contexto sobre el proceso de predicción y permite observar cómo el modelo responde a diferentes entradas.
        return None #Devolvemos None como predicción base para la clase ModeloIA, lo que indica que esta clase no genera una predicción específica y sirve como una base para otras clases que implementarán su propia lógica de predicción.
    def evaluar(self, reales, predichos): #Agregamos un método evaluar a la clase ModeloIA que toma listas de valores reales y predichos, calcula la precisión simulada del modelo comparando ambos conjuntos de datos, y devuelve la precisión calculada, lo que permite evaluar el rendimiento del modelo basado en su precisión simulada establecida.
        if not reales or not predichos: #Si las listas de valores reales o predichos están vacías, devolvemos una precisión de 0.0 para evitar errores de división por cero y proporcionar una evaluación coherente en caso de datos insuficientes.
            return 0.0 #Devolvemos 0.0 si las listas de valores reales o predichos están vacías para evitar errores de división por cero y proporcionar una evaluación coherente en caso de datos insuficientes.
        aciertos = sum(1 for r, p in zip(reales, predichos) if r == p) #Calculamos el número de aciertos comparando los valores reales y predichos utilizando una comprensión de generador, lo que permite contar cuántas predicciones coinciden con los valores reales para evaluar la precisión del modelo.
        self.precision_simulada = aciertos / len(reales) #Calculamos la precisión simulada dividiendo el número de aciertos por la cantidad total de valores reales, lo que proporciona una medida de la precisión del modelo basada en su rendimiento simulado al comparar las predicciones con los valores reales.
        print(f"[{self.nombre}] Precision evaluada: {self.precision_simulada:.2%}") #Agregamos un mensaje que muestra la precisión evaluada del modelo después de comparar los valores reales y predichos, formateada como un porcentaje con dos decimales, lo que proporciona información sobre el rendimiento simulado del modelo basado en su capacidad para generar predicciones precisas.
        return self.precision_simulada #Devolvemos la precisión simulada calculada como resultado de la evaluación, lo que permite obtener una medida cuantitativa del rendimiento del modelo basado en su capacidad para generar predicciones precisas al comparar los valores reales y predichos.
    def __str__(self):
        estado = "Entrenado" if self.entrenado else "No entrenado" #Se determina el estado del modelo en función del atributo entrenado
        return f"ModeloIA: '{self.nombre}' | {estado} | Precision: {self.precision_simulada:.2%}" #Se devuelve una cadena que incluye el nombre del modelo, su estado de entrenamiento y su precisión simulada formateada como porcentaje
    def __add__(self, other): #Se define el método __add__ para permitir la suma de dos objetos de la clase ModeloIA, lo que resultará en un nuevo modelo con una precisión que representa el promedio de ambos modelos
        if not isinstance(other, ModeloIA): #Se verifica que el objeto con el que se está intentando sumar sea una instancia de la clase ModeloIA
            raise TypeError("Solo se pueden sumar objetos ModeloIA") #Si el objeto no es una instancia de ModeloIA, se lanza una excepción indicando que solo se pueden sumar objetos de esa clase
        nombre = f"{self.nombre} ⨁ {other.nombre}" #Se crea un nuevo nombre para el modelo resultante de la suma, combinando los nombres de ambos modelos con un símbolo de suma
        precision = (self.precision_simulada + other.precision_simulada) / 2 #Se calcula la nueva precisión como el promedio de las precisiones simuladas de ambos modelos
        nuevo = ModeloIA(nombre, precision)
        nuevo.entrenado = self.entrenado and other.entrenado #El nuevo modelo se considera entrenado solo si ambos modelos originales están entrenados
        return nuevo
    def __sub__(self, other): #Se define el método __sub__ para permitir la resta de dos objetos de la clase ModeloIA, lo que resultará en un nuevo modelo con una precisión que representa la diferencia entre ambos modelos
        if not isinstance(other, ModeloIA): #Se verifica que el objeto con el que se está intentando restar sea una instancia de la clase ModeloIA
            raise TypeError("Solo se pueden restar objetos ModeloIA") #Si el objeto no es una instancia de ModeloIA, se lanza una excepción indicando que solo se pueden restar objetos de esa clase
        return self.precision_simulada - other.precision_simulada #Se devuelve la diferencia entre las precisiones simuladas de ambos modelos, lo que puede ser útil para comparar su rendimiento
```

# 1.b CLASE HIJA : ClasificadorIA
* Clase hija que hereda de ModeloIA y simula un clasificador.
* Sobrescribe los métodos entrenar() y predecir() con comportamiento específico.
* Sobrescribe entrenar()/ Usamos super() para ejecutar el entrenamiento base, luego añade lógica específica.
* Sobrescribe predecir()/ Que nos permite verifica entrenamiento con super(), luego simula predicción de una clase.
* num_clases / Son cuántas categorías puede distinguir.
* Predicción/ Es cuando se retorna una categoría (ej: "Clase_2")


```python
class ClasificadorIA(ModeloIA):
    def __init__(self, nombre, num_clases): #Se define la clase ClasificadorIA que hereda de ModeloIA y se agrega el parámetro num_clases para especificar el número de clases que el clasificador puede predecir
        super().__init__(nombre) #Es el constructor de la clase padre para inicializar los atributos heredados
        self.num_clases = num_clases #Se asigna el valor del parámetro num_clases al atributo de la clase
    def entrenar(self, datos):
        super().entrenar(datos) #Es el método de la clase padre para realizar el entrenamiento básico
        print(f"[{self.nombre}] Ajustando {self.num_clases} clases") #Agregamos un mensaje indicando que se están ajustando las clases específicas del clasificador
        self.precision_simulada = 0.85 #Se asigna un valor específico a la precisión simulada para el clasificador, lo que puede representar una mejora en comparación con el modelo base
    def predecir(self, entrada):
        super().predecir(entrada) #Metodo de la clase padre para realizar la predicción básica
        pred = f"Clase _ {hash(str(entrada)) % self.num_clases}" #Genera una predicción simulada basada en la entrada y el número de clases, utilizando un hash para distribuir las predicciones entre las clases disponibles
        print(f"[{self.nombre}] Predicción: {pred}") #Agregamos un mensaje indicando la predicción realizada por el clasificador
        return pred #Devuelve la predicción simulada generada para la entrada
```

# 2.b CLASE HIJA : RegresorIA
* La clase hija que hereda de ModeloIA y simulando un regresor.
* Sobrescribe los metodos entrenar() / predecir() que va a tener comportamientos especificos.
* Sobrescribe entrenar() / Usara super() para el entrenamiento base, luego añade optimización de regresión.
* Sobrescribe predecir() / Vamos a verifica un entrenamiento con super(), luego simularemos prediccion numerica continua.


```python
class RegresorIA(ModeloIA):
    def __init__(self, nombre): #Se define la clase RegresorIA que hereda de ModeloIA y no requiere parámetros adicionales en su constructor
        super().__init__(nombre) #Constructor de la clase padre para inicializar los atributos heredados
    def entrenar(self, datos):
        super().entrenar(datos) #Metodo de la clase padre para realizar el entrenamiento básico
        print(f"[{self.nombre}] Optimizando MSE") #Agregamos un mensaje indicando que se está optimizando el error cuadrático medio (MSE), que es una métrica comúnmente utilizada para evaluar regresores
        self.precision_simulada = 0.92 #Asignamos un valor específico a la precisión simulada para el regresor, lo que puede representar una mejora en comparación con el modelo base
    def predecir(self, entrada):
        super().predecir(entrada) #Metodo de la clase padre para realizar la predicción básica
        pred = (hash(str(entrada)) % 100) / 10.0 #Genera una predicción simulada basada en la entrada, utilizando un hash para generar un valor numérico que se ajusta a un rango específico (en este caso, entre 0.0 y 10.0)
        print(f"[{self.nombre}] Predicción: {pred}") #Agregamos un mensaje indicando la predicción realizada por el regresor
        return pred #Devuelve la predicción simulada generada para la entrada, que es un valor numérico representando la salida del regresor para esa entrada específica
```

# 1.2.2 Polimorfismo y Función super()

a) En las clases hijas, usa super() para reutilizar parte del método entrenar() de la clase base y luego añadir comportamiento propio.
Usamos super() dentro de las hijas para llamar al método de la madre y luego añadimos algo extra. Para despues crear una lista con distintos objetos y llamamos al mismo método predecir().

* Polimorfismo / permite una misma instruccion se comporte de forma distinta segun el objeto que recibe para ser mas flexible.

* ModeloIA.entrenar() / nos permite solo entrenamiento base

* ClasificadorIA.entrenar() / nos ayuda en el entrenamiento que se suma a los ajuste de clases

* RegresorIA.entrenar() / junto al entrenamiento se sumara optimización MSE para obtener los resultados.

# 1.2.2.1 POLIMORFISMO
El metodo predecir() se comporta diferente segun el tipo de objeto:
* ModeloBase: devuelve None (predicción base)
* ClasificadorIA: devuelve una clase (ej: "Clase_3")
* RegresorIA: devuelve un número (ej: 7.5)
 Python decide en tiempo de ejecución qué versión del método ejecutar

```python
print("="*50)
print(f"POLIMORFISMO")
print("="*50)
modelo_base = ModeloIA("Base") #Creamos una instancia del modelo base con el nombre "Base"
clasif = ClasificadorIA("Clasificador", 5) #Creamos una instancia del clasificador con el nombre "Clasificador" y especificamos que tiene 5 clases
regres = RegresorIA("Regresor") #Creamos una instancia del regresor con el nombre "Regresor"
datos = [1,2,3,4,5] #Simulamos un conjunto de datos de entrenamiento con una lista de números del 1 al 5
for m in [modelo_base, clasif, regres]: #Iteramos sobre cada modelo creado (modelo_base, clasif y regres) y llamamos al método entrenar para cada uno de ellos, pasando el mismo conjunto de datos de entrenamiento
    m.entrenar(datos) #Entrenamos cada modelo con el mismo conjunto de datos para simular el proceso de entrenamiento y observar cómo cada modelo maneja su propia lógica de entrenamiento, lo que demuestra el concepto de polimorfismo en la programación orientada a objetos
print(f"Predicciones:") #Agregamos un mensaje para indicar que a continuación se mostrarán las predicciones de cada modelo
print("="*50)
for m in [modelo_base, clasif, regres]: #Iteramos nuevamente sobre cada modelo creado y llamamos al método predecir para cada uno de ellos, pasando la misma entrada "dato prueba" para observar cómo cada modelo genera su propia predicción basada en su lógica interna, lo que refuerza el concepto de polimorfismo al permitir que diferentes clases respondan a la misma llamada de método de manera diferente según su implementación específica.
    m.predecir("dato prueba") #Metodo predecir se llama para cada modelo con la misma entrada "dato prueba" para observar cómo cada modelo genera su propia predicción basada en su lógica interna, lo que refuerza el concepto de polimorfismo al permitir que diferentes clases respondan a la misma llamada de método de manera diferente según su implementación específica.
```

# 1.2.4.1  PATRON SINGLETON
* Nos ayuda a garantizar que solo exista una instancia de la clase en todo el codigo para evitar confunciones.
* Con esto no importa cuantas veces se repita GestorSesion()

```python
class GestorSesion:
    _instancia = None #Atributo de clase para almacenar la única instancia de GestorSesion
    def __new__(cls): #Sobrescribimos el método __new__ para controlar la creación de instancias de la clase GestorSesion
        if cls._instancia is None: #Si no existe una instancia de GestorSesion, se crea una nueva instancia utilizando el método __new__ de la clase padre (super()) y se asigna a cls._instancia
            cls._instancia = super().__new__(cls) #Se llama al método __new__ de la clase padre para crear una nueva instancia de GestorSesion y se asigna a cls._instancia
            print("Creando unica instancia.") #Agregamos un mensaje para indicar que se está creando la única instancia de GestorSesion
        return cls._instancia #Se devuelve la instancia única de GestorSesion, ya sea la recién creada o la existente, lo que garantiza que solo haya una instancia de esta clase en todo el programa
    def iniciar_sesion(self): #Agregamos un método para simular el inicio de sesión, que simplemente imprime un mensaje indicando que la sesión ha sido iniciada
        print("Sesión iniciada")
    def cerrar_sesion(self): #Agregamos un método para simular el cierre de sesión, que simplemente imprime un mensaje indicando que la sesión ha sido cerrada
        print("Sesion cerrada")
print("="*50)
print(f"SINGLETON")
print("="*50)
s1 = GestorSesion() #Creamos la primera instancia de GestorSesion, lo que desencadena la creación de la única instancia de esta clase y se imprime el mensaje correspondiente
s2 = GestorSesion() #Intentamos crear una segunda instancia de GestorSesion, pero debido a la implementación del patrón Singleton, se devuelve la misma instancia creada anteriormente y se imprime el mensaje indicando que ya existe una instancia
print("ID s1:", id(s1)) #Imprimimos el ID de la primera instancia (s1) para verificar que es la misma instancia que se creó inicialmente
print("ID s2:", id(s2)) #Imprimimos el ID de la segunda instancia (s2) para verificar que es la misma instancia que se creó inicialmente, lo que confirma que s1 y s2 son referencias a la misma instancia de GestorSesion
s1.iniciar_sesion() #Metodo para iniciar sesión utilizando la instancia s1, lo que simula el inicio de sesión y se imprime el mensaje correspondiente
s2.cerrar_sesion() #Metodo para cerrar sesión utilizando la instancia s2, lo que simula el cierre de sesión y se imprime el mensaje correspondiente, demostrando que ambas variables (s1 y s2) están operando sobre la misma instancia de GestorSesion debido al patrón Singleton implementado en la clase.
```

La herencia nos ayudara para no tener que escribir todo desde cero, ya que la clase ModeloIA ya tiene lo basico (entrenar, predecir, etc.)como tambien ClasificadorIA y RegresorIA que solo agregan / modifican lo necesario en el codigo.
El polimorfismo se ve cuando usamos el mismo método predecir() en todos los modelos, pero cada uno responde diferente. Eso hace que el código sea más ordenado y fácil de usar, porque nos ayuda a tratar todos los modelos de la misma forma aunque hagan cosas distintas.

• ¿Por qué los métodos mágicos son útiles para extender las capacidades de los objetos?

Los métodos mágicos permiten que los objetos se comporten como cosas “naturales” en Python. Por ejemplo, usamos clasif + regres como si fueran números ademas de que estan fusionando modelos.
Permite que el codigo sea más fácil de leer y más intuitivo, para no usar funciones raras, sino a los operadores normales como + o -.

• ¿Cómo ayuda el patrón Singleton a controlar procesos o recursos en sistemas de IA?

El Singleton sirve para que solo exista una instancia de algo importante en todo el codigo,  ya que usamos GestorSesion para asegura que solo haya una sesión activa.
Y nos permitira evita errores como tener varias conexiones al mismo tiempo o duplicar procesos. Ayudandonos a mantener todo bajo control y más organizado.
